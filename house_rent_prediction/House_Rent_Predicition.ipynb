# Data Preprocessing

```python
df.drop("BHK",axis = 1,inplace = True)
```

```python
X = df.drop("Rent",axis = 1)
y = df["Rent"]

cat_col = X.select_dtypes(exclude=np.number)

X = pd.get_dummies(X,columns=cat_col.columns)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
col = X.columns
X = scaler.fit_transform(X)
X = pd.DataFrame(X,columns=col)
```

```python
X
```

# Models

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def mean_cross_value(model,X,y):
    score = cross_val_score(model,X,y,cv=5)
    mean = score.mean()
    return mean
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin = LinearRegression()
```

```python
lin.fit(X_train,y_train)
preds = lin.predict(X_test)
mae_lin = mean_absolute_error(y_test,preds)
rmse_lin = np.sqrt(mean_absolute_error(y_test,preds))
score_lin = lin.score(X_test,y_test)
cv_lin = mean_cross_value(lin,X,y)
r2_lin = r2_score(y_test,preds)
```

## Support Vector Machine



```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit(X_train,y_train)
preds = svr.predict(X_test)
mae_svr = mean_absolute_error(y_test,preds)
rmse_svr = np.sqrt(mean_absolute_error(y_test,preds))
score_svr = svr.score(X_test,y_test)
cv_svr = mean_cross_value(svr,X,y)
r2_svr = r2_score(y_test,preds)
```

## K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
```

```python
knr.fit(X_train,y_train)
preds = knr.predict(X_test)
mae_knr = mean_absolute_error(y_test,preds)
rmse_knr = np.sqrt(mean_absolute_error(y_test,preds))
score_knr = knr.score(X_test,y_test)
cv_knr = mean_cross_value(knr,X,y)
r2_knr = r2_score(y_test,preds)
```

## Gaussian Process

## Tuned XGBoost

```python
from xgboost import XGBRegressor,XGBClassifier
```

```python
xgr = XGBRegressor()
```

```python
param_lst = {
    "learning_rate" : [00.1,0.1,0.15,0.3,0.5],
    "n_estimators" : [100,500,1000,2000,3000],
    "max_depth" : [3,6,9],
    "min_child_weight" : [1,5,10,20],
    "reg_alpha" : [0.001,0.01,0.1],
    "reg_lambda" : [0.001,0.01,0.1]
}
```

```python
xgr = RandomizedSearchCV(estimator=xgr,param_distributions=param_lst,
                            n_iter = 5,scoring="neg_root_mean_squared_error",cv = 5)
```

```python
def mean_cross_value(model,X,y):
    score = cross_val_score(model,X,y,cv=5)
    mean = score.mean()
    return mean
```

```python
xgr.fit(X_train,y_train)
preds_xgbT = xgr.predict(X_test)
mae_xgbT = mean_absolute_error(y_test,preds)
rmse_xgbT = np.sqrt(mean_absolute_error(y_test,preds))
score_xgbT = xgr.score(X_test,y_test)
cv_xgbT = mean_cross_value(xgr,X,y)
r2_xgbT = r2_score(y_test,preds_xgbT)
```

Let's break down each hyperparameter and its corresponding list of values:

1. `"learning_rate"`: This hyperparameter controls the step size at each boosting iteration. The list `[0.01, 0.1, 0.15, 0.3, 0.5]` contains different learning rate values that will be tested.

2. `"n_estimators"`: This hyperparameter determines the number of boosting iterations or the number of trees in the model. The list `[100, 500, 1000, 2000, 3000]` contains different values for the number of estimators to be evaluated.

3. `"max_depth"`: This hyperparameter specifies the maximum depth of each tree in the boosting process. The list `[3, 6, 9]` contains different values for the maximum depth to be considered.

4. `"min_child_weight"`: This hyperparameter defines the minimum sum of instance weights needed in a child node. The list `[1, 5, 10, 20]` contains different values for the minimum child weight to be tested.

5. `"reg_alpha"`: This hyperparameter is the L1 regularization term applied to the weights of the model. The list `[0.001, 0.01, 0.1]` contains different values for the L1 regularization term.

6. `"reg_lambda"`: This hyperparameter is the L2 regularization term applied to the weights of the model. The list `[0.001, 0.01, 0.1]` contains different values for the L2 regularization term.
